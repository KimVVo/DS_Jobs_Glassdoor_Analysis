```python
import pandas as pd
import re 
import matplotlib.pyplot as plt
```

### Reading CSV file

```python
data = pd.read_csv("C:\\Users\\thien\\Jobs Analysis\\Uncleaned_DS_jobs.csv")
data.head()
```

## Cleaning Data


### Salary Estimate

```python
# Replace K by 000 and clean up the unescessary information
data['Salary Estimate'] = data['Salary Estimate'].str.replace('K', '000')
data['Salary Estimate'] = data['Salary Estimate'].apply(lambda x: re.sub(r'[^0-9-]','',x))
data['Salary Estimate']
```

### Job Description and Company Name

```python
# Removing unecessary information on Job Description and Company Name

data['Job Description'] = data['Job Description'].str.replace("\n", "", regex=True)
data['Company Name'] = data["Company Name"].str.split("\n").str[0]
data
```

```python
### Remove unecessary columns

data = data.drop(['Headquarters', 'Sector', 'Competitors'], axis=1)
```

```python
data
```

### Delete rows that don't have enough information

```python
data['Size'] = data['Size'].str.replace('-1', 'NaN')
data
```

```python
data = data[data['Size'] != 'NaN']
data
```

### Removing Founded Column because it is unecessary

```python
# Remove the whole column
data = data.drop(['Founded'], axis=1)
data
```

### Rating

```python
data['Rating'].value_counts()
```

```python
# Replace negative Rating with 0.0

data['Rating'].replace(-1.0, 0.0, inplace=True)
data['Rating'].value_counts()
```

```python
# Find some unknown size value rows
unknown_rows = data[data['Size'] == 'Unknown']
unknown_rows
```

### Type of ownership


```python
replace_dict = {
    '-1': 'NaN',
    'Unknown': 'NaN',
    'Nonprofit Organization': 'Non-profit',
    'Company - Public': 'Public',
    'Private Practice / Firm': 'Private',
    'Company - Private': 'Private',
    'Other Organization': 'Other'
}

data['Type of ownership'] = data['Type of ownership'].replace(replace_dict, regex=True)
```

```python
data['Type of ownership'].value_counts()
```

### Industry

```python
data['Industry'] = data['Industry'].str.replace('-1','NaN')
data['Industry'].value_counts()
```

### Check duplicated rows and delete it if has

```python
duplicates = data.duplicated()
data[duplicates]
```

### Check any null rows and delete it if has

```python
data.isnull().sum()
```

```python
# Export the cleaned data to csv file
data.to_csv('Cleaned_data_science_jobs_glassdoor.csv', index=False)
```

```python
# Extract Salary Estimate to Salary Min and Max, and change to numerical salary value
data['Salary Min'] = data['Salary Estimate'].apply(lambda x: int(x.split('-')[0].replace('$', '').replace(',', '')))
data['Salary Max'] = data['Salary Estimate'].apply(lambda x: int(x.split('-')[1].replace('$', '').replace(',', '')))
```

## DATA ANALYSIS

### Comparing the number of senior roles to non-senior roles


```python
# Find senior roles in Job Title
senior_roles = data['Job Title'].str.contains('senior',flags=re.IGNORECASE, regex=True)
data[senior_roles]
```

```python
# Count cells with 'senior' in 'Job Title' column
senior_roles_num = senior_roles.sum()

# Count cells without 'senior' in 'Job Title' column or can understand Junior role
junior_roles_num = (~data['Job Title'].str.contains('senior', case=False)).sum()

print(f"Senior roles: {senior_roles_num}")
print(f"Junior roles: {junior_roles_num}")
```

```python
# Calculate the percentage between Senior and Junior Roles
total_roles = senior_roles_num + junior_roles_num
senior_percentage = senior_roles_num / total_roles * 100
junior_percentage = junior_roles_num / total_roles * 100

print(f"Senior roles percentage: ({senior_percentage:.2f}%)")
print(f"Junior roles percentage: ({junior_percentage:.2f}%)")
```

```python
# Draw a pie chart to visualize Senior vs Junior Roles Distribution
roles = ['Senior', 'Junior']
sizes = [senior_percentage, junior_percentage]

# Explode a slice if required (e.g., explode the 'Junior Roles' slice)
explode = (0.1, 0)

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=roles, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Adding a title
plt.title('Senior vs Junior Roles Distribution')

# Display the pie chart
plt.show()
```

### The numbers of Jobs per State

```python
# Use .apply() function
def get_city(address):
    return address.split(',')[0]

def get_state(address):
    if ',' in address:
        return address.split(',')[1].strip()
    else:
        return 'Unknown'

data['City'] = data['Location'].apply(lambda x: get_city(x))  
data['State'] = data['Location'].apply(lambda x: get_state(x))  

data.head()
```

```python
grouped_by_state = data.groupby('State')['Job Title'].count()
grouped_by_state
```

```python
# Create a bar plot
# Create the bar plot
plt.figure(figsize=(12, 10))
grouped_by_state.plot(kind='bar')
plt.xlabel('State')
plt.ylabel('Number of Jobs')
plt.title('Distribution of Jobs per State')
plt.show()
```
